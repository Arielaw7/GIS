```python
import geopandas as gpd
import pandas as pd
```

```python
hyogo_ken = gpd.read_file('N03-23_28_230101.geojson')
```

```python
hyogo_ken.head()
```

```python
hyogo_ken.info()
```

```python
hyogo_pop = pd.read_csv('hyogo.csv')
```

```python
hyogo_pop.head()
```

```python
hyogo_pop.info()
```

```python
hyogo_ken.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
hyogo_ken
```

```python
hyogo_ken['KEY_CODE'] = hyogo_ken['KEY_CODE'].astype(int)
```

```python
hyogo_ken.info()
```

```python
hyogo_ken_pop = hyogo_ken.merge(hyogo_pop,on='KEY_CODE')
```

```python
hyogo_ken_pop
```

```python
hyogo_ken_pop.info()
```

```python
hyogo_ken_pop['人口総数'] = hyogo_ken_pop['人口総数'].astype(int)
```

```python
hyogo_ken_pop.info()
```

```python
hyogo_ken_pop.plot(column='人口総数',legend=True)
```
